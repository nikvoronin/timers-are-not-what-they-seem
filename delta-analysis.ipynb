# Timers Analysis 1ms

## System.Timers.Timer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def draw_plot(df, title, mid=True):
    ax = df.plot(title=title)

    ax.set(xlabel="Ticks", ylabel="milliseconds")
    ax.legend(['100% load', 'Idle'])
    ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    if mid:
        ax.axhline(y=df['idle'].median(), xmin=0, xmax=1, color='yellow', linestyle='--', lw=2)
        ax.axhline(y=df['load'].median(), xmin=0, xmax=1, color='blue', linestyle='--', lw=2)
```

```python
timers_idle = pd.read_csv("./data/timer-timers-1ms_idle.csv", delimiter=';', decimal=',')
timers_load = pd.read_csv("./data/timer-timers-1ms_load.csv", delimiter=';', decimal=',')
df_timers = pd.concat([timers_load, timers_idle])
df_timers.info()
```

```python
df_timers.describe()
```

```python
draw_plot(df_timers, "System.Timers.Timer")
```
