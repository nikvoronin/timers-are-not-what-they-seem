## 09 Interval by Stopwatch

```python
import numpy as np
import pandas as pd
```

```python
th_idle = pd.read_csv("../data/th-nosleep-sw-idle.csv", delimiter=';', decimal=',')
th_load = pd.read_csv("../data/th-nosleep-sw-load.csv", delimiter=';', decimal=',')
df_th = pd.concat([th_load, th_idle])
df_th.info()
```

```python
df_th.describe()
```

```python
ax = df_th.plot(title="Interval by Stopwatch", logy=True)

ax.set(xlabel="Ticks", ylabel="ΔTicks, milliseconds")
ax.legend(['100% load', 'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
