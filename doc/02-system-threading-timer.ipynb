# 02 System.Threading.Timer

```python
import numpy as np
import pandas as pd
```

```python
th_idle = pd.read_csv("./data/timer-threading-1ms_idle.csv", delimiter=';', decimal=',')
th_load = pd.read_csv("./data/timer-threading-1ms_load.csv", delimiter=';', decimal=',')
df_threading = pd.concat([th_load, th_idle])
df_threading.info()
```

```python
df_threading.describe()
```

```python
ax = df_threading.plot(title="System.Threading.Timer")

ax.set(xlabel="Ticks", ylabel="milliseconds")
ax.legend(['100% load', 'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.axhline(y=df_threading['idle'].median(), xmin=0, xmax=1, color='yellow', linestyle='--', lw=2)
ax.axhline(y=df_threading['load'].median(), xmin=0, xmax=1, color='blue', linestyle='--', lw=2)
```
