# 03 Multimedia Timer (winmm.dll)

```python
import numpy as np
import pandas as pd
```

```python
mm_idle = pd.read_csv("./data/timer-mm-1ms_idle.csv", delimiter=';', decimal=',')
mm_load = pd.read_csv("./data/timer-mm-1ms_load.csv", delimiter=';', decimal=',')
df_mm = pd.concat([mm_idle, mm_load])
df_mm.info()
```

```python
df_mm.describe()
```

```python
ax = df_mm.plot(title="Multimedia Timer (winmm.dll)")

ax.set(xlabel="Ticks", ylabel="milliseconds")
ax.legend(['100% load', 'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.axhline(y=df_mm['idle'].median(), xmin=0, xmax=1, color='yellow', linestyle='--', lw=2)
ax.axhline(y=df_mm['load'].median(), xmin=0, xmax=1, color='blue', linestyle='--', lw=2)
```
