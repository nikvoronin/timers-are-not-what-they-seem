# 04 TwinCAT 3. Soft Real-Time System. Notifications

```python
import numpy as np
import pandas as pd
```

```python
tc3_idle = pd.read_csv("../data/tc3-1ms_idle.csv", delimiter=';', decimal=',')
tc3_load = pd.read_csv("../data/tc3-1ms_load.csv", delimiter=';', decimal=',')
df_tc = pd.concat([tc3_load, tc3_idle])
df_tc.info()
```

```python
df_tc.describe()
```

```python
ax = df_tc.plot(title="TwinCAT 3 Notifications")

ax.set(xlabel="Ticks", ylabel="Î”Ticks, milliseconds")
ax.legend(['100% load', 'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.axhline(y=df_tc['idle'].median(), xmin=0, xmax=1, color='yellow', linestyle='--', lw=2)
ax.axhline(y=df_tc['load'].median(), xmin=0, xmax=1, color='blue', linestyle='--', lw=2)
```
