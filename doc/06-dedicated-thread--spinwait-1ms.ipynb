## 06 SpinWait 1ms

```python
import numpy as np
import pandas as pd
```

```python
sw_idle = pd.read_csv( "../data/th-spin-1ms_idle.csv", delimiter=';', decimal=',')
sw_load = pd.read_csv( "../data/th-spin-1ms_load.csv", delimiter=';', decimal=',')
sw_load9 = pd.read_csv("../data/th-spin-1ms_load9.csv", delimiter=';', decimal=',')
df_sw = pd.concat([sw_load, sw_load9, sw_idle])
df_sw.info()
```

```python
df_sw.describe()
```

```python
ax = df_sw.plot(title="SpinWait 10000us")

ax.set(xlabel="Ticks", ylabel="ΔTicks, milliseconds")
ax.legend(['100% load', '90% load',  'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
