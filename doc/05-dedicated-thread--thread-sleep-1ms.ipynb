# Dedicated Thread
## 05 Thread.Sleep 1ms

```python
import numpy as np
import pandas as pd
```

```python
slp_idle = pd.read_csv("./data/th-sleep-1ms_idle.csv", delimiter=';', decimal=',')
slp_load = pd.read_csv("./data/th-sleep-1ms_load.csv", delimiter=';', decimal=',')
slp_load9 = pd.read_csv("./data/th-sleep-1ms_load9.csv", delimiter=';', decimal=',')
df_slp = pd.concat([slp_load, slp_load9, slp_idle])
df_slp.info()
```

```python
df_slp.describe()
```

```python
ax = df_slp.plot(title="Thread.Sleep 1ms")

ax.set(xlabel="Ticks", ylabel="milliseconds")
ax.legend(['100% load', '90% load',  'Idle'], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
